```python
import random
import string
import pandas as pd
from datetime import datetime, timedelta

# Function to generate a random Ethereum-like address
def generate_random_address():
    return "0x" + ''.join(random.choices(string.hexdigits.lower(), k=40))

# Function to generate random transaction ID
def generate_random_transaction_id():
    return "0x" + ''.join(random.choices(string.hexdigits.lower(), k=64))

# Function to generate random token from a predefined list
def generate_random_token():
    tokens = ["ETH", "USDT", "DAI", "USDC", "WBTC"]
    return random.choice(tokens)

# Function to generate a random status
def generate_random_status():
    statuses = ["success", "failed", "pending"]
    return random.choice(statuses)

# Function to generate random amount between 50 and 500
def generate_random_amount():
    return round(random.uniform(50, 500), 2)

# Function to generate a random timestamp within a given range
def generate_random_timestamp(start_date, end_date):
    delta = end_date - start_date
    random_minutes = random.randint(0, delta.total_seconds() // 60)
    random_timestamp = start_date + timedelta(minutes=random_minutes)
    return random_timestamp.strftime("%Y-%m-%dT%H:%M:%SZ")

# Generate 1001 rows of random data
data = []
start_date = datetime(2025, 1, 1, 12, 0, 0)
end_date = datetime(2025, 1, 1, 18, 0, 0)

for _ in range(1001):
    transaction_id = generate_random_transaction_id()
    timestamp = generate_random_timestamp(start_date, end_date)
    from_address = generate_random_address()
    to_address = generate_random_address()
    amount = generate_random_amount()
    token = generate_random_token()
    status = generate_random_status()

    data.append({
        "transaction_id": transaction_id,
        "timestamp": timestamp,
        "from_address": from_address,
        "to_address": to_address,
        "amount": amount,
        "token": token,
        "status": status
    })

# Create a DataFrame
df_random = pd.DataFrame(data)

# Save the DataFrame as a CSV
file_path_random = '/mnt/data/random_starknet_transactions_1001.csv'
df_random.to_csv(file_path_random, index=False)

# Check the number of rows
print("Number of rows in the dataset:", len(df_random))
print("Dataset saved to:", file_path_random)

# If running in Google Colab, use this line to download the file to your local machine
from google.colab import files
files.download(file_path_random)
```
